```
import pandas as pd
from google.colab import drive
drive.mount('/content/gdrive')

from pandas import DataFrame, read_csv
import matplotlib.pyplot as plt

from bs4 import BeautifulSoup

movie_2017 = pd.read_csv('/content/gdrive/My Drive/Colab_Data/Doc/2017_01_movie.csv', header=3, thousands=',' , encoding='euc-kr')


movie_2017.rename(columns={movie_2017.columns[4] : "매출액점유율"}, inplace=True)


movie_2017.drop([0], inplace=True)
movie_2017.drop([493], inplace=True)

movie_2017['흥행지수'] = (movie_2017['누적매출액 '] / movie_2017['누적관객수 ']) * movie_2017['상영횟수 ']
movie_2017.sort_values(by='흥행지수', ascending=False).head(5)

movie_2017['흥행순위'] = movie_2017['흥행지수'].rank(ascending=False)
movie_2017.sort_values(by='흥행순위', ascending=True).head()
```
